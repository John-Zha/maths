# 1

```python
df = pd.read_json("box_office_2025.json")
df.info()
```

# 2

```python
def parse_percentage(value):
    if value in ('-','<0.1%'):
        return 0.0
    return float(value.replace('%',''))

df['domestic_pct']=df['domestic_share'].map(parse_percentage)
df['foreign_pct']=df['foreign_share'].map(parse_percentage)

df.head(),df.info()
```

# 5

```python
top_domestic=df.sort_values('domestic_pct',ascending=False)
top_foreign=df.sort_values('foreign_pct',ascending=False)

top_domestic.head()
#top_foreign.head()
```

```python
ranked = df[["title", "worldwide_gross"]].copy()
ranked['gross_rank']=ranked['worldwide_gross'].rank(ascending=False).astype(int)
ranked['domestic_rank']=df['domestic_pct'].rank(ascending=False).astype(int)

ranked
```
